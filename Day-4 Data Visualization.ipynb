```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,23,34,45,49,56,61,68,71,78])

y = np.array([23,31,38,43,49,56,65,71,79,85])

z = np.array([98,89,86,76,71,68,61,54,49,34])
```

## Line Plot

```python
plt.plot(x, y);
```

```python
plt.plot(x, z);
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', color = 'blue',
        mfc = 'r', ls = '--', ms = 7);

plt.grid()
# marker => <, >, ^, o, *, H
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = '*', color = 'cyan',
        mfc = 'blue', ls = '--', ms = 7);

plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', color = 'blue',
        mfc = 'r', ls = '--', ms = 7);

plt.plot(x, z, marker = 'o', color = 'orange',
        mfc = 'blue', ls = '--', ms = 7);

plt.grid()
```

## Subplot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', color = 'blue',
        mfc = 'r', ls = '--', ms = 7);

plt.grid()


plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = '*', color = 'cyan',
        mfc = 'blue', ls = '--', ms = 7);

plt.grid()
```

## Bar Graph

```python
students = [56, 65, 46, 51, 61]

courses = ['Java', 'Python', 'R', 'Kotlin',
           'Dart']
```

```python
plt.title('Bar Plot')
plt.xlabel('Courses')
plt.ylabel('No. of Students')
plt.bar(courses, students, color = 'black',
       edgecolor = 'red', width = 0.6,
       linewidth = 3, alpha = 0.7);
```

## Scatter Plot

```python
plt.scatter(x, y, marker= '*');
plt.grid()
```

## Pie Chart

```python
plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct = '%.2f%%',
       shadow= True, radius = 0.8, labeldistance= 1.2,
       colors = ['cyan', 'orange', 'skyblue',
                'purple', 'red'], startangle= 30,
       explode = [0,0,0.2,0,0]);
```

```python
plt.figure(figsize=(16, 12))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', color = 'blue',
        mfc = 'r', ls = '--', ms = 7);

plt.plot(x, z, marker = 'o', color = 'orange',
        mfc = 'blue', ls = '--', ms = 7);

plt.grid()

plt.subplot(2,2,2)
plt.title('Bar Plot')
plt.xlabel('Courses')
plt.ylabel('No. of Students')
plt.bar(courses, students, color = 'black',
       edgecolor = 'red', width = 0.6,
       linewidth = 3, alpha = 0.7);
plt.grid()

plt.subplot(2,2,3)
plt.scatter(x, y, marker= '*');
plt.grid()

plt.subplot(2,2,4)
plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct = '%.2f%%',
       shadow= True, radius = 0.8, labeldistance= 1.2,
       colors = ['cyan', 'orange', 'skyblue',
                'purple', 'red'], startangle= 30,
       explode = [0,0,0.2,0,0]);

plt.savefig('MY PLOT')
```

## Box Plot

```python
marks = [67, 63, 70, 71, 72, 56, 51, 21, 95]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)

q1 = np.percentile(marks, 25)
```

```python
q3, q1
```

```python
iqr = q3 - q1
iqr  # interquartile range
```

```python
# lower Bound
lb = q1 - 1.5 * iqr

# upper bound

up = q3 + 1.5 * iqr
```

```python
lb, up
```

## Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python

sns.jointplot(x = 'tip', y = 'total_bill',
            data = tips, hue = 'time');
plt.show()
```

```python
# histogram

plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'], kde = True)

plt.show()
```

```python
plt.title('Outliers in tip')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxenplot(x = 'day', y = 'tip',
             data = tips);

plt.show()
```
